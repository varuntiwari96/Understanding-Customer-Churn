```python
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score


dt_search_space = {
    'criterion': hp.choice('criterion', ['gini', 'entropy']),
    'max_depth': hp.choice('max_depth', [None, hp.quniform('max_depth_int', 1, 100, 1)]),
    'min_samples_split': hp.uniform('min_samples_split', 0.1, 1.0),
    'min_samples_leaf': hp.uniform('min_samples_leaf', 0.1, 0.5),
    'max_features': hp.choice('max_features', [None, 'sqrt', 'log2'])  # Valid options for max_features
}


# the objective function for Decision Tree classifier
def dt_objective_function(params):
    
    max_depth = int(params['max_depth']) if params['max_depth'] is not None else None
    
    clf_params = {
        'criterion': params['criterion'],
        'max_depth': max_depth,
        'min_samples_split': params['min_samples_split'],
        'min_samples_leaf': params['min_samples_leaf'],
        'max_features': params['max_features']
    }
    clf = DecisionTreeClassifier(**clf_params)
    cv_score = cross_val_score(clf, X_train, y_train, cv=5, scoring='roc_auc').mean()
    return {'loss': -cv_score, 'status': STATUS_OK}

# Run the optimization for Decision Tree classifier
print("Optimizing hyperparameters for Decision Tree...")
trials = Trials()
best_dt = fmin(dt_objective_function, dt_search_space, algo=tpe.suggest, max_evals=50, trials=trials)
print("Best hyperparameters for Decision Tree:", best_dt)
```
